# Matplotlib

```python
import matplotlib.pyplot as plt
#numpy module. Fill your code
import numpy as np

xpoints=np.array([1,2,3,4])
ypoints=np.array([2,4,6,8])

#Plot these points without drawing a line
#Your code here
plt.plot(xpoints, ypoints, 'o')
plt.title("Plot without Line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
#Plotting with marker: Plot these points with a marker(Star marker)
#Your code here
plt.plot(xpoints, ypoints, '*')
plt.title("Plot with Star Marker")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
#Using fmt format, add circular marker,red color and Dashed line
#Your code here
plt.plot(xpoints, ypoints, 'o--r')
plt.title("Plot with Circular Marker, Red Color, and Dashed Line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

#Add xlabel,ylabel and title for the plot.
#Your code here
plt.plot(xpoints, ypoints, 'o')
plt.title("Plot with Labels")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

#Create a scatter plot for xpoints and ypoints
#Your code here
plt.scatter(xpoints, ypoints)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
#Set color to the scatter plot. Blue,Green,Red and yellow color for each point respectively
colors = ['blue', 'green', 'red', 'yellow']
plt.scatter(xpoints, ypoints, color=colors)
plt.title("Scatter Plot with Colors")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import random
import numpy as np

#Set the seed of random to 20
#Your code here
random.seed(20)

arr1=np.array([1,24,31,45,73,81,94,25])

#Using the random module pick 4 different random numbers from arr1 and return their sum.
#Your code here
random_indices = random.sample(range(len(arr1)), 4)
random_numbers = arr1[random_indices]
random_sum = sum(random_numbers)

print(f"Selected random numbers: {random_numbers}")
print(f"Sum of selected random numbers: {random_sum}")
```
