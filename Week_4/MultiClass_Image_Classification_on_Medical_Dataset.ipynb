```python
import os
import cv2
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from PIL import Image
from sklearn.metrics import accuracy_score, roc_auc_score, average_precision_score, f1_score
from tqdm import tqdm
```

```python
# Define paths to CSV files and image directories
train_csv_path = 'KCDH2024_Training_GroundTruth(in).csv'
test_csv_path = 'KCDH2024_Test_GroundTruth(in).csv'
lesion_csv_path = 'KCDH2024_Training_LesionGroupings(in).csv'
train_image_dirs = ['HAM10000_images_part_1', 'HAM10000_images_part_2']
test_image_dir = 'KCDH2024_Test_Input'

# Load CSV files
train_df = pd.read_csv(train_csv_path)
test_df = pd.read_csv(test_csv_path)
lesion_df = pd.read_csv(lesion_csv_path)
```

```python
# Preprocess image using OpenCV
def preprocess_image(image_path):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (224, 224))  # Resize to match model input size
    image = cv2.GaussianBlur(image, (5, 5), 0)  # Apply Gaussian blur
    return image

# Custom Dataset class
class DermoscopyDataset(Dataset):
    def __init__(self, dataframe, image_dirs, transform=None):
        self.dataframe = dataframe
        self.image_dirs = image_dirs
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        img_name = self.dataframe.iloc[idx, 0] + ".jpg"
        for image_dir in self.image_dirs:
            img_path = os.path.join(image_dir, img_name)
            if os.path.exists(img_path):
                image = preprocess_image(img_path)
                image = Image.fromarray(image)
                break
        else:
            raise FileNotFoundError(f"Image {img_name} not found in any provided directories.")
        
        labels = self.dataframe.iloc[idx, 1:].values.astype('float32')
        
        if self.transform:
            image = self.transform(image)
        
        return image, torch.tensor(labels)
```

```python
# Transformations
data_transforms = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Create datasets and dataloaders
train_dataset = DermoscopyDataset(train_df, train_image_dirs, transform=data_transforms)
test_dataset = DermoscopyDataset(test_df, [test_image_dir], transform=data_transforms)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
# Model setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = models.resnet50(pretrained=True)
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 7)
model = model.to(device)

criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# Training function
def train_model(model, criterion, optimizer, dataloader, num_epochs=25):
    model.train()
    for epoch in range(num_epochs):
        running_loss = 0.0
        for inputs, labels in tqdm(dataloader, desc=f"Epoch {epoch+1}/{num_epochs}"):
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item() * inputs.size(0)
        
        epoch_loss = running_loss / len(dataloader.dataset)
        print(f'Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}')
```

```python
# Evaluation function
def evaluate_model(model, dataloader):
    model.eval()
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, labels in tqdm(dataloader, desc="Evaluating"):
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            outputs = model(inputs)
            preds = torch.sigmoid(outputs).cpu().numpy()
            all_preds.append(preds)
            all_labels.append(labels.cpu().numpy())
    
    all_preds = np.vstack(all_preds)
    all_labels = np.vstack(all_labels)
    
    accuracy = accuracy_score(np.argmax(all_labels, axis=1), np.argmax(all_preds, axis=1))
    auc = roc_auc_score(all_labels, all_preds, average='macro')
    map_score = average_precision_score(all_labels, all_preds, average='macro')
    f1 = f1_score(np.argmax(all_labels, axis=1), np.argmax(all_preds, axis=1), average='macro')
    
    print(f'Accuracy: {accuracy:.4f}, AUC: {auc:.4f}, Mean AP: {map_score:.4f}, F1 Score: {f1:.4f}')
    
    return accuracy, auc, map_score, f1
```

```python
# Train and evaluate the model
train_model(model, criterion, optimizer, train_loader, num_epochs=5)
```

```python
evaluate_model(model, test_loader)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, precision_recall_curve

# Function to plot loss curve
def plot_loss_curve(loss_values, title='Loss Curve'):
    plt.figure(figsize=(10, 6))
    plt.plot(range(len(loss_values)), loss_values, marker='o', linestyle='-', color='b')
    plt.title(title)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.grid(True)
    plt.show()

# Function to plot confusion matrix
def plot_confusion_matrix(cm, class_names, title='Confusion Matrix'):
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.title(title)
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()

# Function to plot ROC curves
def plot_roc_curves(y_true, y_score, class_names):
    plt.figure(figsize=(12, 8))
    for i, class_name in enumerate(class_names):
        fpr, tpr, _ = roc_curve(y_true[:, i], y_score[:, i])
        plt.plot(fpr, tpr, label=f'ROC curve (class {class_name})')

    plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc='lower right')
    plt.show()

# Function to plot Precision-Recall curves
def plot_precision_recall_curves(y_true, y_score, class_names):
    plt.figure(figsize=(12, 8))
    for i, class_name in enumerate(class_names):
        precision, recall, _ = precision_recall_curve(y_true[:, i], y_score[:, i])
        plt.plot(recall, precision, label=f'Precision-Recall curve (class {class_name})')

    plt.title('Precision-Recall Curve')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.legend(loc='lower left')
    plt.show()

# Training function with loss tracking
def train_model(model, criterion, optimizer, dataloader, num_epochs=25):
    model.train()
    loss_values = []
    for epoch in range(num_epochs):
        running_loss = 0.0
        for inputs, labels in tqdm(dataloader, desc=f"Epoch {epoch+1}/{num_epochs}"):
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item() * inputs.size(0)
        
        epoch_loss = running_loss / len(dataloader.dataset)
        loss_values.append(epoch_loss)
        print(f'Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}')
    
    return loss_values

# Evaluation function with confusion matrix and ROC/PR curves
def evaluate_model(model, dataloader, class_names):
    model.eval()
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, labels in tqdm(dataloader, desc="Evaluating"):
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            outputs = model(inputs)
            preds = torch.sigmoid(outputs).cpu().numpy()
            all_preds.append(preds)
            all_labels.append(labels.cpu().numpy())
    
    all_preds = np.vstack(all_preds)
    all_labels = np.vstack(all_labels)
    
    # Compute metrics
    accuracy = accuracy_score(np.argmax(all_labels, axis=1), np.argmax(all_preds, axis=1))
    auc = roc_auc_score(all_labels, all_preds, average='macro')
    map_score = average_precision_score(all_labels, all_preds, average='macro')
    f1 = f1_score(np.argmax(all_labels, axis=1), np.argmax(all_preds, axis=1), average='macro')
    
    print(f'Accuracy: {accuracy:.4f}, AUC: {auc:.4f}, Mean AP: {map_score:.4f}, F1 Score: {f1:.4f}')
    
    # Confusion Matrix
    cm = confusion_matrix(np.argmax(all_labels, axis=1), np.argmax(all_preds, axis=1))
    plot_confusion_matrix(cm, class_names)
    
    # ROC Curves
    plot_roc_curves(all_labels, all_preds, class_names)
    
    # Precision-Recall Curves
    plot_precision_recall_curves(all_labels, all_preds, class_names)
    
    return accuracy, auc, map_score, f1

# Train and evaluate the model
loss_values = train_model(model, criterion, optimizer, train_loader, num_epochs=25)
plot_loss_curve(loss_values)

# List of class names
class_names = ['MEL', 'NV', 'BCC', 'AKIEC', 'BKL', 'DF', 'VASC']

accuracy, auc, map_score, f1 = evaluate_model(model, test_loader, class_names)
```
